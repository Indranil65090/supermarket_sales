```python
fig = plt.figure()  
ax = plt.axes()  
  
x = np.linspace(0, 10, 1000)  
ax.plot(x, np.sin(x))  

x = np.arange(0.0, 2, 0.01)  
y1 = np.sin(2 * np.pi * x)  
y2 = 1.2 * np.sin(4 * np.pi * x)  
fig, ax = plt.subplots(1, sharex=True)  
ax.plot(x, y1, x, y2, color='black')  
ax.fill_between(x, y1, y2, where=y2 >= y1, facecolor='blue', interpolate=True)  
ax.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)  
ax.set_title('fill between where')  
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=1000,n_features=2,centers=4,cluster_std=3.0,random_state=42)
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap="rainbow")
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)  # Adjust the number of clusters as needed
km.fit(data[0])

f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10, 6))

ax1.scatter(data[0][:, 0], data[0][:, 1], c=km.labels_, cmap="rainbow")
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap="rainbow")

plt.show()
```
